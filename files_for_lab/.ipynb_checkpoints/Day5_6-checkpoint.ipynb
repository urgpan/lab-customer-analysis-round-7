```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import math
import warnings
warnings.filterwarnings('ignore')
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer, StandardScaler, OrdinalEncoder
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv(r'C:\Users\urgpa\OneDrive\Escritorio\Documentos\RELHACKERS\LEWORK\DAY3\LAB\lab-customer-analysis-round-3\files_for_lab\csv_files\marketing_customer_analysis.csv')
```

# CLEANING

```python
def clean_makerting_customer_analysis (x):
    #RENAME COLUMNS
    x = x.rename(columns={'EmploymentStatus':'Employment Status'})#rename typos in columns
    #x['Effective To Date'] = x['Effective To Date'].astype('datetime64[ns]')#datatype
    #DROP UNWANTED COLUMNS
    customer = x['Customer']
    x = x.drop(['Customer','Sales Channel','Number of Policies','Employment Status','Vehicle Class','Number of Open Complaints','Months Since Policy Inception', 'Months Since Last Claim'], axis = 1)
    x = x.drop('Effective To Date', axis = 1)
    x = x.drop('Gender', axis = 1)
    x = x.drop('State', axis = 1)
    
    return x, customer
```

```python
data_clean , customer_data = clean_makerting_customer_analysis(raw_data)
```

```python
data_clean.isna().sum()
```

```python
#SPLIT CATEGORICAL AND NUMERICAL
data_num = data_clean.select_dtypes(include = np.number)#save numerical data
data_cat = data_clean.select_dtypes(include = np.object)#save categorical data
```

```python
#NAN or ZERO VALUES REPLACE BY MEDIAN in numerical
for i in data_num.columns:  
    median = data_num[i].median()
    data_num.loc[data_num[i] == 0, i] = median
   
    
```

```python
print(data_num.max())
```

```python
correlations_matrix = data_clean.corr()
mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(correlations_matrix, mask=mask, annot=True)
plt.show()
```

```python
for i in data_num.columns:
    sns.boxplot(x=data_num[i])
    plt.show()
```

# PROCESSING

```python
#OUTLINERS
for u in data_num.columns:
    iqr = np.percentile(data_num[u],75) - np.percentile(data_num[u],25) #interquartile range
    print(iqr)
    upper_limit = np.percentile(data_num[u],75) + 0.5*iqr #whysker positions
    print(upper_limit)
    lower_limit = np.percentile(data_num[u],25) - 0.5*iqr
    print(lower_limit)
    data_num_out = data_num[(data_num[u]>lower_limit) & (data_num[u]<upper_limit)]#filter data through whyskers
    print(u)
```

```python
data_num_out.shape
```

```python
for i in data_num_out.columns:
    sns.boxplot(x=data_num_out[i])
    plt.show()
```

```python
data_clean.shape
```

```python
#NORMALIZE NUMERICAL DATA
norm = Normalizer()
optimusprime = norm.fit(data_num_out)#fit the transformer to optimusprime
x_normalized = optimusprime.transform(data_num_out)#transform and normalize with the fit
data_num_norm = pd.DataFrame(x_normalized)
```

```python
#COLUMN NAMES!
data_num_norm.columns = data_num_out.columns
```

```python
data_cat.columns
```

```python
#SPLIT INTO CARDINAL AND ORDINAL
data_cat_ord = data_cat[['Coverage','Education','Vehicle Size','Renew Offer Type','Policy Type','Policy']]
data_cat_car = data_cat.drop(['Coverage','Education','Vehicle Size','Renew Offer Type','Policy Type','Policy'],axis = 1)
```

```python
encoder = OrdinalEncoder()
# transform data
data_ord_encoded = encoder.fit_transform(data_cat_ord)
data_cat_encoded = pd.DataFrame(data_ord_encoded)
data_cat_encoded.columns = data_cat_ord.columns
```

```python
#DUMMYFY CATEGORICAL DATA
data_cat_dummy = pd.get_dummies(data_cat_car)
#MAYBE SEPARETE INTO ORDINAL AND CARDINAL
```

```python
data_cat.head()
```

```python
#CONCAT
#rest indexes before concateanitigng
data_num_norm.reset_index(drop = True ,inplace = True)
data_cat_encoded.reset_index(drop = True ,inplace = True)

concat_data = pd.concat([data_num_norm,data_cat_encoded,data_cat_dummy], axis = 1)
```

```python
concat_data = concat_data.fillna(concat_data.median())
```

```python
concat_data.isna().sum()
```

```python
concat_data.dtypes
```

```python
#DROP TARGET X Y
Y = concat_data['Total Claim Amount'] #variable u want to predict
data_targeted = concat_data.drop('Total Claim Amount', axis = 1) #save the new file with the dropped data
X = data_targeted
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=100)
```

```python
X_train.dtypes
```

```python
X_train.isna().sum()
```

```python
lm = linear_model.LinearRegression()

# We "train" our linear model based on the data available on X_train and y_train
model = lm.fit(X_train,y_train)
```

```python
# We make predictions on X_test
#predictions  = lm.predict(X_test)
```

```python
predictions = lm.predict(X_test)
print("R2 value is = ",round(r2_score(y_test, predictions),2))
print("The intercept of the model is = ",lm.intercept_)
print("The coefficients of the model are = ",lm.coef_)
predictions = lm.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print("The mse of the model is = ",mse)
```
